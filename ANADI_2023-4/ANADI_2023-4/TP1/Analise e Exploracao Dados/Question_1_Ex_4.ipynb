```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
dados = pd.read_csv("CO_data.csv" , delimiter=",")
dados.columns = ["country","year","population","gdp","cement_co2","co2","coal_co2","energy_per_capita","energy_per_gdp","flaring_co2","gas_co2","methane","nitrous_oxide","oil_co2"]
dados_portugal = dados.loc[dados['country'] == 'Portugal']

#Ex 4
dados_maior_que_1999 = dados.loc[dados['year'] > 1999, ['country','year', 'coal_co2']]
years_new = dados_maior_que_1999['year']
paises = ['United States', 'China', 'India', 'European Union (27) (GCP)', 'Russia']
dados_pais = {}
cores = {
    'United States': 'blue',
    'China': 'orange',
    'India': 'green',
    'European Union (27) (GCP)': 'red',
    'Russia': 'purple'
}
for pais in paises:
    dados_pais[pais] = dados_maior_que_1999[dados_maior_que_1999['country'] == pais]

for pais in paises:
    plt.plot(dados_pais[pais]['year'],dados_pais[pais]['coal_co2'],label=pais, color=cores[pais])

plt.title('Evolução da emissão de Co2 por carvão em diferentes países no período de 2000-2021')
plt.xlabel('Ano')
plt.ylabel('Emissao Co2 por carvão')
plt.legend()
plt.show()
```
