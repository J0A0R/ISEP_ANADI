```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
dados = pd.read_csv("CO_data.csv" , delimiter=",")
dados.columns = ["country","year","population","gdp","cement_co2","co2","coal_co2","energy_per_capita","energy_per_gdp","flaring_co2","gas_co2","methane","nitrous_oxide","oil_co2"]
dados_portugal = dados.loc[dados['country'] == 'Portugal']

#Ex 2
emissoes_co2 = ['cement_co2','coal_co2','flaring_co2','gas_co2','methane','nitrous_oxide','oil_co2']   
cores = {
    'cement_co2': 'blue',
    'coal_co2': 'orange',
    'flaring_co2': 'green',
    'gas_co2': 'red',
    'methane': 'purple',
    'nitrous_oxide': 'black',
    'oil_co2': 'pink'
}


for emissao in emissoes_co2:
    plt.plot(dados_portugal['year'],dados_portugal[emissao],label=emissao, color=cores[emissao])
    
plt.title('Evolução dos diferentes tipos de emissões no período de 1900-2021')
plt.xlabel('Ano')
plt.ylabel('Emissao CO2')
plt.legend()
plt.show()
```
