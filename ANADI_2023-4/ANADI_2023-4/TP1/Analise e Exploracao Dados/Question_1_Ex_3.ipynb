```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
dados = pd.read_csv("CO_data.csv" , delimiter=",")
dados.columns = ["country","year","population","gdp","cement_co2","co2","coal_co2","energy_per_capita","energy_per_gdp","flaring_co2","gas_co2","methane","nitrous_oxide","oil_co2"]
dados_portugal = dados.loc[dados['country'] == 'Portugal']

#Ex 3
dados_espanha = dados.loc[dados['country'] == 'Spain']
co2_capita_portugal = dados_portugal['co2'] / dados_portugal['population']
co2_capita_espanha = dados_espanha['co2'] / dados_espanha['population']
plt.plot(dados_portugal['year'],co2_capita_portugal,label='Portugal',color="blue")
plt.plot(dados_portugal['year'],co2_capita_espanha,label='Espanha',color="red")
plt.xlabel('Ano')
plt.ylabel('Valores per capita dos pases')
plt.legend()
plt.title('Co2 per capita em espanha e portugal no perodo 1900-2021')
plt.show()
```
