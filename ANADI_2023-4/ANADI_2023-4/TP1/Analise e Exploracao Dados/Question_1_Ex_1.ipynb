```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
dados = pd.read_csv("CO_data.csv" , delimiter=",")
dados.columns = ["country","year","population","gdp","cement_co2","co2","coal_co2","energy_per_capita","energy_per_gdp","flaring_co2","gas_co2","methane","nitrous_oxide","oil_co2"]

#Ex 1
dados_portugal = dados.loc[dados['country'] == 'Portugal']
emissoes_totais_co2 = dados_portugal['cement_co2'] + dados_portugal['co2'] + dados_portugal['coal_co2'] + dados_portugal['flaring_co2'] + dados_portugal['gas_co2'] + dados_portugal['oil_co2']
dados_portugal.insert(14, "emissoes_totais_co2", emissoes_totais_co2)
plt.plot(dados_portugal['year'],dados_portugal['emissoes_totais_co2'])
plt.xlabel('Ano')
plt.ylabel('Emissões Totais de CO2')
plt.title('Emissões de Co2 em Portugal no período 1900-2021')
max_emissoes = dados_portugal['emissoes_totais_co2'].max()
years_max_emissoes = dados_portugal[dados_portugal['emissoes_totais_co2'] == max_emissoes]['year']
print("Years with maximum total CO2 emissions:", years_max_emissoes.values)
plt.show()
```
